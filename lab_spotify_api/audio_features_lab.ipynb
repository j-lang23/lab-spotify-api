```python
def search_song(title, artist):
    query = f"tracks:{title} artist: {artist}"
    try:
        results = sp.search(q=query)
        tracks = results['tracks']['items']
        track_id = tracks[0]['id']
        return f"spotify:track:{track_id}"
    except:
        print(f"No results found for {title} by {artist}")
        return ''
```

```python
import pandas as pd
import time

def get_audio_features(track_ids):
   
    features = {
        'danceability': [],
        'energy': [],
        'key': [],
        'loudness':[],
        'mode': [],
        'speechiness': [],
        'acousticness': [],
        'instrumentalness': [],
        'liveness': [],
        'valence': [],
        'tempo': [],
        'type': [],
        'id': [],
        'uri': [],
        'track_href': [],
        'analysis_url': [],
        'duration_ms': [],
        'time_signature': [],
        'open_url':[]}
    
    for track_id in track_ids:
        try:
            results = sp.audio_features(track_id)
            if results is None: 
                continue 
            for key in list(results[0].keys()):
                features[key].append(results[0][key])
            features['open_url'].append("https://open.spotify.com/track/"+results[0]['id'])
        except Exception as e:
            print(f"Error getting audio features for track_id {track_id}: {str(e)}")
    
    df = pd.DataFrame(features)
    return df
```

```python
dfs = []
for i in range(0, top_100.shape[0], 50):
    print("Getting the audio features for chunk ", i)
    list_of_track_ids = top_100.iloc[i:i+50, :]['track_id'].tolist()
    print("Sleeping some time...")
    time.sleep(20)
    df_chunk = get_audio_features(list_of_track_ids)
    dfs.append(df_chunk)

df = pd.concat(dfs, ignore_index=True)
```
